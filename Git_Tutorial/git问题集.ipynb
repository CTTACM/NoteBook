### Q2:

   在上面的操作之后，把git bash设置为utf-8，中文，然后git add，git commit ，git push到远程仓库，却发现远程仓库里面出现了两个FockPrice文件夹，一个文件夹里面有ARIMA模型文档.ipynb文件，一个文件夹理有神经网络模型价格预测.ipynb文件。怎么回事？

**原因:**

   我在push的时候使用了-u参数，即`git push -u origin master`命令，这个命令只是在第一次推送master分支所有内容的时候才使用。我们这次只是修改而不是第一次推送，正确的命令应该是`git push origin master`，用于推送最新修改。

**解决办法**

使用`git push origin master`命令推送最新修改。

## Q3

一次添加文件的时候，在本地文件夹ForkPrice下创建了一个.ipynb文件，然后add，commit，push，但是在push的时候加了参数-u，结果打开github网站发现出现
了两个ForkPrice文件，一个里面是原有内容，一个里面是.ipynb文件。然后查看本地仓库发现只有一个文件夹，而且github也没有直接删除文件夹的按钮，那么我
该怎么删除这个远程的文件夹呢？
**原因**

通过观察，终于发现，两个文件夹是不一样的，一个是ForkPrice，一个是Forkprice，其中P的大小写不同。

**解决办法**

我的解决办法就是新建一个Fork_Price文件夹，然后里面装好内容，提交到远程库。然后再把远程库中两个错误的文件夹删掉。
